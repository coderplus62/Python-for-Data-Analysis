```python
import pandas as pd
import seaborn as sns
air_quality = pd.read_pickle('air_quality.pkl')
```

```python
air_quality.columns
```

```python
sns.displot(data=air_quality,x='TEMP',hue='quarter',col='station')
```

```python
sns.catplot(data=air_quality,x='quarter',y='TEMP',hue='prior_2016_ind',kind='box')
```

```python
sns.catplot(data=air_quality,x='quarter',y='TEMP',hue='prior_2016_ind',kind='bar')
```

```python
sns.catplot(data=air_quality,x='PM2.5_category',y='TEMP',kind='strip',hue='quarter')
# strip is the default setting
```

```python
sns.pairplot(data=air_quality[['TEMP','PRES','PM2.5','quarter']],hue='quarter')
```

```python
sns.relplot(data=air_quality,x='TEMP',y='PRES',hue='quarter')
```

```python
sns.relplot(data=air_quality,x='TEMP',y='PRES',hue='quarter',col='year',col_wrap=3)
```

```python
sns.relplot(data=air_quality,x='TEMP',y='PRES',hue='quarter',col='year',col_wrap=3,style='quarter')
```

```python
sns.relplot(data=air_quality,x='TEMP',y='PRES',hue='quarter',col='year',col_wrap=3,size='PM2.5')
```
