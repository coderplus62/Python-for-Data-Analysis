```python
import pandas as pd
import seaborn as sns
air_quality = pd.read_pickle('air_quality.pkl')
```

```python
sns.displot(data=air_quality,y='TEMP')
```

```python
sns.displot(data=air_quality,x='TEMP')
sns.displot(data=air_quality,x='TEMP',binwidth=5)
sns.displot(data=air_quality,x='TEMP',bins=5)
sns.displot(data=air_quality,x='TEMP',bins=5,height=3,aspect=3) # aspect * height = width 
```

```python
sns.displot(data=air_quality,x='TEMP',kind='kde')
sns.displot(data=air_quality,x='TEMP',kde=True)
sns.displot(data=air_quality,x='TEMP',kde=True,rug=True)
```

```python
sns.catplot(data=air_quality,y='TEMP',kind='box')
sns.catplot(data=air_quality,x='TEMP',kind='box')
```

```python
sns.catplot(data=air_quality,x='TEMP_Category',kind='count')
sns.catplot(data=air_quality,x='PM2.5_category',kind='count')
sns.catplot(data=air_quality,x='PM2.5_category',kind='count',aspect=2.5)
```

```python
sns.catplot(data=air_quality,x='wd',kind='count',aspect=2.5)
```
