```python
import pandas as pd
air_quality = pd.read_pickle('air_quality.pkl')
```

```python
air_quality.head()
```

```python
air_quality.groupby(by='PM2.5_category').groups
```

```python
air_quality.groupby(by='PM2.5_category').groups.keys()
```

```python
air_quality['PM2.5_category'].head(20)
```

```python
air_quality.groupby(by='PM2.5_category').get_group('Good')
```

```python
air_quality.sort_values('date_time').groupby(by='year').first()
```

```python
air_quality.sort_values('date_time').groupby(by='year').last()
```

```python
air_quality.groupby(by='TEMP_Category').size()
```

```python
air_quality['TEMP_Category'].value_counts(sort=False)
```

```python
air_quality.groupby(by='quarter').mean()
```

```python
air_quality.groupby(by='quarter')[['PM2.5','TEMP']].mean()
# air_quality.groupby(by='quarter').mean()[['PM2.5','TEMP']]
# air_quality[['PM2.5','TEMP','quarter']].groupby(by='quarter').mean() from O'Relly
```

```python
air_quality.groupby(by='quarter')[['PM2.5','TEMP']].describe()
```

```python
air_quality.groupby(by='quarter')[['PM2.5','TEMP']].agg(['min','max'])
```

```python
air_quality.groupby(by='day_of_week_name')[['PM2.5','TEMP','RAIN']].agg({'PM2.5':['min','max','mean'],'TEMP':'mean','RAIN':'mean'})
```

```python
air_quality.groupby(by=['quarter','TEMP_Category'])[['PM2.5','TEMP']].mean()
```

```python
air_quality.groupby(by=['TEMP_Category','quarter'])[['PM2.5','TEMP']].mean()
```

```python
air_quality.groupby(['year','quarter','month'])['TEMP'].agg(['min','max','mean'])
```
