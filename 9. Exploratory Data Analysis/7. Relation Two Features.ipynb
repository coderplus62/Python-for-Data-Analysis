# PART 1

```python
import pandas as pd
import seaborn as sns
air_quality = pd.read_pickle('air_quality.pkl')
```

```python
sns.relplot(data=air_quality,x='TEMP',y='PRES',kind='scatter')
```

```python
sns.pairplot(data=air_quality[['PM2.5','TEMP','PRES']])
```

```python
sns.jointplot(data=air_quality,x='TEMP',y='PRES')
```

```python
air_quality.corr()
```

<img src='correlation_coefficient.jpg'>

```python
sns.heatmap(data=air_quality.corr(),cmap='coolwarm')
```
