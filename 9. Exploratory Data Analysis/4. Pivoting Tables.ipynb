```python
import pandas as pd
student = pd.read_csv('student.csv')
```

```python
student.info()
```

```python
student
```

```python
pd.pivot_table(student,
               index=['sex','internet'],
               values='score')
```

```python
pd.pivot_table(student,
               index=['sex','internet'],
               values='score',
               aggfunc=['min','mean','max'])
```

```python
pd.pivot_table(student,
               index=['sex','internet'],
               values='score',
               aggfunc='mean',
               columns='studytime')
```

```python
pd.pivot_table(student,
               index=['sex','internet'],
               values=['score','age'],
               aggfunc='mean',
               columns=['studytime'],
               fill_value=-999)
```

```python
pd.pivot_table(student,
               index=['sex'],
               values=['score','age'],
               aggfunc='mean',
               columns=['internet','studytime'],
               fill_value=-999)
```

```python
pd.pivot_table(student,
               index=['familysize'],
               values=['score'],
               aggfunc='mean',
               columns=['sex'])
```

```python
pd.pivot_table(student,
               index=['familysize'],
               values=['score'],
               aggfunc='mean',
               columns=['sex'],
               margins=True,
               margins_name='Average score total')
```

```python
student[student['sex']=='F'].mean()
```

```python
pd.pivot_table(student,
               index=['studytime'],
               aggfunc={'age':['max','min'],
                        'score':'median'},
               columns=['sex'])
```

```python
pd.pivot_table(student,
               index=['studytime'],
               aggfunc= {'score':lambda x:x.max()-x.min()},
               columns=['sex'])

# pd.pivot_table(student,
#                index=['studytime'],
#                columns=['sex'],
#                values=['age','score'],
#                aggfunc= lambda x:x.max()-x.min())
```
