```python
import pandas as pd
wdi = pd.read_pickle('wdi.pkl')
num_cols = wdi.select_dtypes(include='number').columns
```

```python
wdi.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
```

```python
from sklearn.impute import IterativeImputer
```

```python
iter_imp = IterativeImputer(min_value=wdi[num_cols].min(),max_value=wdi[num_cols].max())
```

```python
wdi[num_cols].min()
```

```python
wdi[num_cols].max()
```

```python
imputed_num = iter_imp.fit_transform(wdi[num_cols])
```

```python
wdi[num_cols]=imputed_num
```
