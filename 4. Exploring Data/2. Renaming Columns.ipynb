# Renaming Columns

```python
import pandas as pd
```

```python
hero_power = pd.read_pickle('Hero Power.pkl')
marvel_hero = pd.read_pickle('Marvel Hero.pkl')
dc_hero = pd.read_pickle('DC Hero.pkl')
```

```python
hero_info = pd.concat([marvel_hero,dc_hero],ignore_index=True)
```

```python
heroes = pd.merge(hero_info,hero_power,left_on='name',right_on='hero_names')
```

```python
heroes.columns = heroes.columns.str.lower()
```

```python
heroes.rename(columns={'hero_names':'hero names'})
```

```python
heroes = heroes.rename(columns={'hero_names':'hero names'})
```

```python
heroes = heroes.rename(columns=str.title)
```

```python
hero = heroes.rename(mapper=str.lower,axis=1)
hero.head(2)
```
