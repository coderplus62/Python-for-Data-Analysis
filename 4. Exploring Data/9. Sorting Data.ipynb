```python
import pandas as pd
```

```python
hero = pd.read_pickle('hero_8.pkl')
```

```python
hero.head()
```

```python
hero.sort_values(by='race',ascending=False)
```

```python
hero.sort_values(by='bmi',ascending=False).head()
```

```python
hero[['name','height','race']].sort_values(by=['height','race'])
```

```python
hero[['name','height','race']].sort_values(by=['race','height'],ascending=[True,False])
```

```python
hero[['name','height']].sort_values(by='height',ascending=False).head(10)
```

```python
hero.nlargest(n=10,columns='height')
```

```python
hero.nsmallest(10,'height')
```
